<p style="font-size:300%;font-weight:300;padding:20px;text-align:center;margin-top:0.75in;">
        <span style="font-weight:700;color:#d17000;">1.</span> randframe.gaussframe</p>
<p style="line-height:35px;font-size:150%;margin:1in;margin-top:0px;">I made this function because I was tired of creating random data to test on from scratch. <strong>Data scientists will at some point need to create random data, and that is what gaussframe() does.</strong> It does so by picking values on a normal distribution with a given mean and standard deviation. By DEFAULT, the mean is 50 and the standard deviation is 25. These are arbitrary, so one can also pass <span style="color:purple">mu</span>=<span style="color:forestgreen">[]</span> or <span style="color:purple">sigma</span>=<span style="color:forestgreen">[]</span> to the function as keyword arguments. You specify the average and/or standard deviation for every column index, or j for j in (input)<sub>ij</sub>. This function is <span style="color:purple">column-centric</span>, so our index convention is <span style="color:purple">column_index, row_index</span> (or j,i) instead of row then column.</p>

```python
df = gaussframe(5,10)
print('>>>gaussframe(5,10)\n')
df
```

```python
df.plot()
```

```python
print("\n>>>gaussframe(3,30,names=['AAPL','F','CAKE'],chop=2,mu=[78,372,210],sigma=[3,23,42])\n")
data = gaussframe(3,30,names=['AAPL','F','CAKE'],chop=2,mu=[78,372,210],sigma=[3,23,42])
print('Summary Stats:')
data.describe()
```

<p style="line-height:35px;font-size:150%">Notice <strong>the correlation between the values in the mean and std column of the summary stats and the arguments in our calling of gaussframe().</strong><br><br>Now, let's look at all the data and <em>see how the summary manifests in the entries.</em></p>

```python
data
```

```python
data.plot()
```

<p style="font-size:300%;font-weight:300;text-align:center;margin-top:0.75in;">
        <span style="font-weight:700;color:#d17000;">2.</span> benford.benfordness</p>

<p style="line-height:35px;font-size:150%;margin:1in;margin-top:0px;">
We saw how <strong>gaussframe()</strong> can return a pandas.DataFrame that meets certain optional specification using a normal distribution. <strong>The benfordness() function is a tool to help you spot randomly generated data or otherwise suspicious anamolies.</strong> <a href="https://en.wikipedia.org/wiki/Benford%27s_law#:~:text=Benford's%20law%2C%20also%20called%20the,life%20sets%20of%20numerical%20data.&text=If%20the%20digits%20were%20distributed,about%2011.1%25%20of%20the%20time.">Benfords Law</a> states that there is a certain probability that for digits 1-9, the probability of being the leading digit in an entry decreases going from 1 to 9.<br><br><strong>The Probabilities:</strong><br>{<span style="font-weight:700;color:#d17000;">1</span> : 30.1, <span style="font-weight:700;color:#d17000;">2</span> : 17.6, <span style="font-weight:700;color:#d17000;">3</span> : 12.5 , <span style="font-weight:700;color:#d17000;">4</span> : 9.7, <span style="font-weight:700;color:#d17000;">5</span> : 7.9, <span style="font-weight:700;color:#d17000;">6</span> : 6.7, <span style="font-weight:700;color:#d17000;">7</span> : 5.8, <span style="font-weight:700;color:#d17000;">8</span> : 5.1, <span style="font-weight:700;color:#d17000;">9</span> : 4.6}<br><br> This function checks and visualizes the 'benfordness' of a list of data, optimized for a pandas.Series object.<br>
<strong>Below, you will notice that our randomly generated data sticks out when compared with the benford curve of digit frequencies.</strong></p>

```python
print(">>>benfordness(data['AAPL'])\n")
benfordness(data['AAPL'])
```

```python
print(">>>benfordness(data['F'])\n")
benfordness(data['F'])
```

```python
print(">>>benfordness(data['CAKE'])\n")
benfordness(data['CAKE'])
```

<p style="line-height:35px;font-size:150%">Let's use gaussframe() and attempt to <strong>create a series of data that will pass the benfordness test</strong> with a mean absolute error of less than %5.</p>

```python
test = gaussframe(1, 30, names=['test_series'], mu=[498,], sigma=[495,])
test['test_series'] = abs(test['test_series'])
print(">>>test = gaussframe(1, 30, names=['test_series'], mu=[498,], sigma=[495,])\n>>>benfordness(test['test_series'])\n")
benfordness(test['test_series'])
```

        <span style="font-weight:700;color:#d17000;">3.</span> grant_regression.lr</p>

<p style="line-height:35px;font-size:150%;margin:1in;margin-top:0in;">Grant Fisher and I were able to collaborate on <strong>a linear regression function that uses matrix operations</strong> on the backend, so the first key feature of this function is its ability to display the matrix system of linear equations that creates the line of best fit by passing in <span style="color:purple">show_matrix</span>=<span style="color:forestgreen">True</span> to the function. Additionally, one can also add annotations to the input point(s) by passing in <span style="color:purple">annotate</span>=<span style="color:forestgreen">True</span>.</p>

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 5, 12, 19, 26, 34, 41]
lr(x=x, y=y, annotate=False, show_matrix=False)
```

```python
#Using the dataframe created by an instance of gaussframe() from above.
print(">>>lr(x=df.index,y=df['Variable 4'],annotate=False, show_matrix=False)\n")
lr(x=df.index,y=df['Variable 4'],annotate=False, show_matrix=False)
```

```python
#Using ordered (x,y) couples separated by commas as positional arguments.
print(">>>lr((234,234),(234,223),(72,324), show_matrix=True, annotate=True)\n")
lr((234,234),(234,223),(72,324), show_matrix=True, annotate=True)
```

```python
#Changing the Linear regression by changing the Points.
#Mixing gaussframe and lr.
print(">>>dumbf = gaussframe(2,30,names=['Time','Depth'],chop=2,mu=[100,200],sigma=[100,20])\n")
print(">>>lr(x=list(dumbf['Time']),y=list(dumbf['Depth']),show_matrix=False,annotate=False)\n")
dumbf = gaussframe(2,30,names=["Time","Depth"],chop=2,mu=[100,200],sigma=[100,20])
lr(x=list(dumbf['Time']),y=list(dumbf['Depth']),show_matrix=False,annotate=False)
```

```python
quick_cube = {
    #define x domain
    'domain' : list(range(1,21)),
    'output': [i**3 for i in range(1,21)],
}
lr(x=quick_cube['domain'], y=quick_cube['output'], annotate=True)
```

```python
#Passing list comprehensions to lr with kwargs x and y.
#{
#x=[i for i in range(20)],
#y=[1/(i+0.1) for i in range(20)],
#}

lr(x=[i for i in range(20)],y=[1/(i+0.5) for i in range(20)],show_matrix=False)
```
